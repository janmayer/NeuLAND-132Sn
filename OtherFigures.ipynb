```python
%matplotlib

# with silicon grease
sg_xs = [60, 85, 110, 135, 165, 190, 215, 235]
sg_ys = [130, 129, 127, 125, 126, 131, 136, 139]
sg_yerr = 8 * [2]

# without
ng_xs = [60, 85, 110, 135, 165, 190, 215, 235]
ng_ys = [144, 143, 146, 146, 152, 149, 151, 155]
ng_yerr = 8 * [2]

fig = plt.figure(figsize=(6.224, 6.224 / 1.618))

ax = fig.add_subplot(111)
ax.set_xlim(0, 250)
ax.set_ylim(120, 170)

ax.errorbar(sg_xs, sg_ys, yerr=sg_yerr, fmt="o-", linewidth=0.5, label="with optical coupling")
ax.errorbar(ng_xs, ng_ys, yerr=ng_yerr, fmt="s:", linewidth=0.5, label="no optical coupling")
ax.axvline(250 / 2, color="black", linestyle=":", linewidth=0.5)

ax.set_xlabel(r"Position along scintillator bar [cm]", horizontalalignment="right", x=1)
ax.set_ylabel(r"Time resolution $\sigma$ [ps]", horizontalalignment="right", y=1)

ax.legend(loc="upper left", prop={"size": "x-small"})

fig.tight_layout()
fig.savefig("results/opticalglue.pdf", bbox_inches="tight")
display(fig)
```
